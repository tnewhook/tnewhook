```python
# URL of the webpage to scrape
url = "https://www.worldometers.info/geography/alphabetical-list-of-countries/"

# Send a request to the webpage
response = requests.get(url)
response.raise_for_status()  # Ensure we notice bad responses

# Parse the webpage content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the list of countries
table = soup.find('table')

# Extract the list of countries
countries = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    if columns:
        country = columns[1].get_text(strip=True)
        countries.append(country)

# Print the list of countries
# for country in countries:
#     print(country)

Countries = pd.DataFrame({'Country':countries})
```

```python
def match_country(coffee_name, countries):
    result =  'None'
    for country in countries['Country']:
        if country.lower() in coffee_name.lower():
#             return country
            result = country  
        elif 'java' in coffee_name.lower():
            result = 'Indonesia'
        elif 'sumatra' in coffee_name.lower():
            result = 'Indonesia'
        elif 'flores' in coffee_name.lower():
            result = 'Indonesia'
        elif 'sulawesi' in coffee_name.lower():
            result = 'Indonesia'
        elif 'timor leste' in coffee_name.lower():
            result =  'Timor Leste'
        elif 'ethiopique' in coffee_name.lower():
            result = 'Ethiopia'
    return result
#         print('Result: ' + str(result))
    

coffeeLocationData['Regions'] = coffeeLocationData['Regions'].str.replace(' and ','')
coffeeLocationData['Regions'] = coffeeLocationData['Regions'].str.replace(' - ',',')
coffeeLocationData['Regions'] = coffeeLocationData['Regions'].str.replace(', ',',')
coffeeLocationData['Regions'] = coffeeLocationData['Regions'].apply(lambda x: x.strip())

# Apply the function to add a new column
```
